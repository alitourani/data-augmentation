```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import os
%matplotlib inline

# Main Parameters and Variabls
imageFolderDirectory = "E:/Datasets/Dataset 9 - the Cars (Stanford)/Train"
numberOfInputImages = 5
singleImagePath = ""

# Let's read some images from the directory
imageList = os.listdir(imageFolderDirectory)
count = 0
for imageFileName in imageList:
    if count < numberOfInputImages:
        singleImagePath = imageFolderDirectory + "/" + imageFileName
        image = imageio.imread(singleImagePath)
        print("Original image #" + str(count) + " - " + imageFileName + ":")
        ia.imshow(image)
    count = count + 1
```
